```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/placement.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.drop(columns=['Unnamed: 0']) #preprocessing: dropping the 'Unnamed: 0' column
```

```python
df.head()
```

```python
#EDA
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])
```

```python
# Extract the input and output columns
X=df.drop(columns=['placement'])
Y=df['placement']
```

```python
X
```

```python
Y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
#sacle the value
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
#Evaluating The model
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
clf.fit(X_train,Y_train)
```

```python
Y_pred = clf.predict(X_test)
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score = accuracy_score(Y_test,Y_pred)
print('Accuracy score of the training data : ', accuracy_score)
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# .values converts Pandas DataFrames/Series into NumPy arrays
plot_decision_regions(X_train, Y_train.values, clf=clf, legend=2)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Regions')
plt.show()
```

```python
#Export the model
import pickle
pickle.dump(clf,open('model.pkl','wb'))
```
